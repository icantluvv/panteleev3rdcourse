```python
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
text = ''' Творчество А. С. Пушкина прекрасно! Оно до сих пор вдохновляет людей, будит в них добро, помогает «родиться» поэтам и прозаикам.

Я хотела бы побывать в чудесном и загадочном Лукоморье. Собственными глазами увидеть русалку в ветвях раскидистого дуба, настоящих богатырей и колдунов, погладить ученого кота, послушать его увлекательные сказки и народные песни, погрузиться в мир чудес.

Познакомиться с Бабой-Ягой, которая хоть и страшная, но все равно кажется доброй и родной. А уж прокатиться в ее ступе, которая «идет, бредет сама собой» и побывать в избушке на курьих ножках «без окон, без дверей» — мечта детства.

Неведомые дорожки с невиданными следами зверей пробудили во мне исследователя и любителя бродить по реальным дорогам и полям, рассматривать, кто какие оставляет следы.

«Лукоморье» и другие сказки Александра Сергеевича открыли для меня мир чудес и загадок, научили находить них намеки и уроки.

Например, как нужно себя вести, какое поведение неприемлемо и оттолкнет от меня друзей.

Прекрасное Лукоморье пробудило во мне сказочника. Я полюбила придумывать сказочные истории. '''
```

```python
words = word_tokenize(text)
words
```

```python
len(words)
```

```python
stop = stopwords.words('russian')
stop
```

```python
word_filtered = [word for word in words if word not in stop]
len(word_filtered)
```

```python
pr = len(word_filtered)/len(words)
pr
```
